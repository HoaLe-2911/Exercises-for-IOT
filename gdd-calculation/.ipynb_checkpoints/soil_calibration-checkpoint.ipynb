Part 1: Calibration Data Collection
I collected 4 soil samples with varying moisture levels. Below are the recorded sensor readings along with the wet and dry weights (dried at 110°C) for each sample:

Sample 1 (Very Dry): Sensor Reading = 850 | Wet Weight = 110g | Dry Weight = 105g

Sample 2 (Slightly Damp): Sensor Reading = 550 | Wet Weight = 125g | Dry Weight = 105g

Sample 3 (Wet): Sensor Reading = 350 | Wet Weight = 140g | Dry Weight = 105g

Sample 4 (Waterlogged): Sensor Reading = 200 | Wet Weight = 155g | Dry Weight = 105g

Part 2: Gravimetric Soil Moisture Calculation
Using the formula: Gravimetric Soil Moisture (%) = ((W_wet - W_dry) / W_dry) * 100

Sample 1: (110 - 105) / 105 * 100 = 4.76%

Sample 2: (125 - 105) / 105 * 100 = 19.04%

Sample 3: (140 - 105) / 105 * 100 = 33.33%

Sample 4: (155 - 105) / 105 * 100 = 47.61%

Part 3 & 4: Calibration Graph and Calibrated Reading

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Experimental Data
sensor_readings = np.array([850, 550, 350, 200])
moisture_percent = np.array([4.76, 19.04, 33.33, 47.61])

# 2. Plotting the Calibration Graph
plt.figure(figsize=(10, 6))
plt.scatter(sensor_readings, moisture_percent, color='blue', label='Calibration Samples', zorder=5)

# Calculate and plot the line of best fit
m, b = np.polyfit(sensor_readings, moisture_percent, 1)
plt.plot(sensor_readings, m*sensor_readings + b, color='red', linestyle='--', label='Line of Best Fit')

# Formatting the graph
plt.title('Soil Moisture Sensor Calibration Graph')
plt.xlabel('Sensor Reading (0-1023)')
plt.ylabel('Gravimetric Soil Moisture (%)')
plt.gca().invert_xaxis() # Invert X-axis since a lower reading indicates higher moisture
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# 3. Print conclusion and calibrated reading
print("--- CALIBRATION RESULT ---")
print(f"Calibration Equation: Moisture (%) = {m:.4f} * Sensor Reading + {b:.4f}")
print("Conclusion: Based on the line of best fit, we can calculate the actual moisture for any sensor reading.")
print("For example, if the sensor returns a reading of 500, the gravimetric soil moisture is approx {:.1f}%".format(m*500 + b))
```
